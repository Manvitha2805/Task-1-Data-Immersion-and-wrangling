Check and Handling Missing Values

//identifying missing values

```python
df.isnull().sum()
```

//remove missing values

```python
df = df.dropna()
df.isnull().sum()
```

// identifying duplicate values

```python
df.duplicated().sum()
```

// drop the duplicates

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

Convert Data Types

//convert date column to datetime format

```python
df["Date"] = pd.to_datetime(df["Date"])
```

//convert Quantity and Amount to numeric format

```python
df["Qty"]=pd.to_numeric(df["Qty"],errors="coerce")
df["Amount"]=pd.to_numeric(df["Amount"],errors="coerce")
```

Data set exploring after changing datatypes

FEATURE ENGINEERING

//Extract Year,Month,Day,
Day of Week , Day of Name from Date

```python
df["Year"]= df["Date"].dt.year
df["Month"]= df["Date"].dt.month
df["Day"]= df["Date"].dt.day
df["Day of Week"]= df["Date"].dt.dayofweek
df["Day Name"]= df["Date"].dt.day_name()
```

// create total revenue column

```python
df["Total_Revenue"]=df["Qty"]*df["Amount"]
```
